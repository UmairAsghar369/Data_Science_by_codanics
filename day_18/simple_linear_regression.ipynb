# Machine Learning
### 1- Simple Linear Regression

```python
import pandas as pd

df = pd.read_csv('Salary_dataset.csv')

df = df.drop(columns=['Unnamed: 0'])
df.head()
```

### Step-2 Splitting data into trainging data and splitting data

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
y.head()
```

```python
# import liberaries for model splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Step-3 Fit Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
model
```

### Step-4 ploting

```python
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, color='blue')

plt.plot(X_train, model.predict(X_train), color='green')  # regression line
plt.title('Training set')
plt.xlabel('Years of Experience')   
plt.ylabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color='blue')

plt.plot(X_test, model.predict(X_test), color='green')  # regression line
plt.title('Test set')
plt.xlabel('Years of Experience')   
plt.ylabel('Salary')
plt.show()
```

### Step-5 Testing or Evaluating your model

```python
# model fitness
model.score(X_test, y_test) # score of testing data
```

```python
model.score(X_train, y_train) # score of training data
```

### Step-6 Prediction of unknown values

```python
model.predict([[5]]) # jiska 5 saal ka experience hai uska salary kitna hoga
```

```python
model.predict(X_test) # test data ke salary ki prediction karna)
```
